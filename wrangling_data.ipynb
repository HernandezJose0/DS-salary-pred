## Wrangling data

```python
import pandas as pd
import numpy as np
from unidecode import unidecode
from datetime import date, timedelta
```

```python
jobs_df=pd.read_csv("jobs.csv")
jobs_df.head()
```

```python
jobs_df=jobs_df[jobs_df["Salary"].isnull()!=True]
jobs_df["Yearly"]=jobs_df["Salary"].apply(lambda x: 1 if 'al año' in x.lower() else 0)
jobs_df["Hourly"]=jobs_df["Salary"].apply(lambda x: 1 if 'por hora' in x.lower() else 0)
jobs_df.head()
```

```python
salary=jobs_df["Salary"].apply(lambda x: x.split('al')[0]).apply(lambda x: x.split('por')[0])
salary.head()
```

```python
rm_char=salary.apply(lambda x: x.replace('$','').replace(',',''))
rm_char.head()
```

```python
jobs_df["Min_Salary"]=rm_char.apply(lambda x:int(x.split('-')[0]))
jobs_df["Max_Salary"]=rm_char.apply(lambda x: int(x.split('-')[1]) if '-' in x else int(x))
jobs_df["Avg_Salary"]=(jobs_df.Max_Salary+jobs_df.Min_Salary)/2
jobs_df.tail()
```

```python
jobs_df["Title"]=jobs_df["Title"].apply(lambda x:x.replace('nuevo empleo',''))
jobs_df["Remote"]=jobs_df["Location"].apply(lambda x:1 if 'remoto' in x.lower() else 0)
jobs_df["State"]=jobs_df["Location"].apply(lambda x:x.split(',')[1] if ',' in x else x).apply(lambda x:x.split('+')[0] if '+' in x else x)
jobs_df["Posted_Date"]=jobs_df["Date"].apply(lambda x:unidecode(x.lower().split('hace')[-1]).replace('dias','').replace('+','')\
.replace('mas de','').replace('dia','') if 'hace' in x.lower() else 0).apply(lambda x:date.today()-timedelta(int(x)))
jobs_df.tail()
```

```python
jobs_df["Python"]=jobs_df["Description"].apply(lambda x:1 if 'python' in x.lower() else 0)
print(jobs_df.Python.value_counts())
jobs_df["R_studio"]=jobs_df["Description"].apply(lambda x:1 if 'R ' in x else 0)
print(jobs_df.R_studio.value_counts())
jobs_df["Spark"]=jobs_df["Description"].apply(lambda x:1 if 'spark' in x.lower() else 0)
print(jobs_df.Spark.value_counts())
jobs_df["Excel"]=jobs_df["Description"].apply(lambda x:1 if 'excel' in x.lower() else 0)
print(jobs_df.Excel.value_counts())
jobs_df["SQL"]=jobs_df["Description"].apply(lambda x:1 if 'sql' in x.lower() else 0)
print(jobs_df.SQL.value_counts())
jobs_df["Hadoop"]=jobs_df["Description"].apply(lambda x:1 if 'hadoop' in x.lower() else 0)
print(jobs_df.Hadoop.value_counts())
```

```python
def title_cat(title):
    if 'data scientis' in title.lower() or 'cientifico de datos' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower() or 'ingeniero de datos' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower() or 'analista' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'ML'
    else:
        return 'na'
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'manager' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'junior'
    else:
        return 'na'
```

```python
jobs_df["title_cat"]=jobs_df["Title"].apply(title_cat)
jobs_df.title_cat.value_counts()
```

```python
jobs_df["level"]=jobs_df["Title"].apply(seniority)
jobs_df.level.value_counts()
```

```python
jobs_df.State.value_counts()
```

```python
jobs_df["State"]=jobs_df["State"].apply(lambda x: 'CDMX' if 'polanco' in x.lower() else x.strip()).apply(lambda x:'Remoto' if 'remoto' in x.lower() else x)\
.apply(lambda x:'CDMX' if 'ciudad de' in x.lower() else x).apply(lambda x:'B.C.' if 'baja' in x.lower() else x)\
.apply(lambda x:'Méx.' if 'mexico' in unidecode(x.lower()) else x.strip())
jobs_df.State.value_counts()
```

```python
jobs_df["Min_Salary"]=jobs_df.apply(lambda x:x.Min_Salary*8*22 if x.Hourly == 1 else x.Min_Salary,axis=1)
jobs_df["Min_Salary"]=jobs_df.apply(lambda x:x.Min_Salary/12 if x.Yearly==1 else x.Min_Salary,axis=1)
jobs_df["Max_Salary"]=jobs_df.apply(lambda x:x.Max_Salary*8*22 if x.Hourly == 1 else x.Max_Salary,axis=1)
jobs_df["Max_Salary"]=jobs_df.apply(lambda x:x.Max_Salary/12 if x.Yearly==1 else x.Max_Salary,axis=1)
jobs_df["Avg_Salary"]=(jobs_df.Max_Salary+jobs_df.Min_Salary)/2
```

```python
jobs_df["desc_len"]=jobs_df["Description"].apply(lambda x:len(x))
jobs_df.to_csv("jobs_cleaned.csv",index=False)
```
